```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')
```

```python
n_dpi = 480
p = (10,10)
p_color = (0,0,0)
n_img = 1

x1 = 0
x2 = 20

ms = [1, 0.9, 0.8, 0.7, 0.6]
ns = [0, 1, 2, 3, 4]
colors = [(24/255, 192/255, 196/255), (246/255, 33/255, 150/255), (162/255, 103/255, 245/255), (243/255, 144/255, 126/255), (255/255, 228/255, 107/255)]

largura = 14/2.54
altura = 10/2.54
```

```python
plt.figure(figsize=(largura, altura))
plt.plot(p[0], p[1], marker='o', color=p_color)
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(x1,x2)
plt.ylim(x1,x2)
plt.savefig(f'img/ani/ani0{n_img}.png', dpi=n_dpi)
n_img += 1
```

```python
plt.figure(figsize=(largura, altura))
for i in range(0,1):
    plt.plot([x1,x2], [x1*ms[i]+ns[i],x2*ms[i]+ns[i]], color=colors[i], label=f'$y={ms[i]}x+{ns[i]}$')
plt.plot(p[0], p[1],'o', color=p_color)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.xlim(x1,x2)
plt.ylim(x1,x2)
plt.savefig(f'img/ani/ani0{n_img}.png', dpi=n_dpi)
n_img += 1
```
