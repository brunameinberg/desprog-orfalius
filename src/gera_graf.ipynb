```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
from matplotlib.ticker import FuncFormatter

plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')
plt.rcParams['axes.facecolor'] = '#dfdfea'
plt.rcParams['axes.formatter.use_locale'] = True
```

```python
def cria_pasta(nome_pasta):
    diretorio_atual = os.getcwd()
    caminho_pasta = os.path.join(diretorio_atual, 'img', nome_pasta)
    os.path.exists(caminho_pasta)

def formatar_com_virgulas(valor, pos):
    return "{:,.0f}".format(valor)
```

```python
SALVAR_GRAFICOS = True
N_DPI = 240
P_COLOR = (0,0,0)

COLORS = [(24/255, 192/255, 196/255), (246/255, 33/255, 150/255), (162/255, 103/255, 245/255), (243/255, 144/255, 126/255), (255/255, 228/255, 107/255)]

LARGURA = 14/2.54
ALTURA = 10/2.54
```

```python
nome_img = 'pontos'
n = 1
fig, axs = plt.subplots(1, 1, figsize=(LARGURA, ALTURA))
c = [(0, 0, 0), (1, 1, 1)]
for i in range(0, int(14*n)):
    for j in range(0, int(10*n)):
        axs.plot((i + 0.5)/n, (j + 0.5)/n, marker='o', markersize=3.5/n, color=c[random.randint(0, 1)])
axs.set_xlabel('$X$')
axs.set_ylabel('$Y$')
axs.set_xlim(0, 14)
axs.set_ylim(0, 10)
axs.xaxis.set_major_locator(MultipleLocator(1/n))
axs.yaxis.set_major_locator(MultipleLocator(1/n))
axs.set_xticklabels([])
axs.set_yticklabels([])

plt.savefig(f'img/{nome_img}.png', dpi=N_DPI) if SALVAR_GRAFICOS else None
```

```python
nome_img = 'inf_retas'

tangentes = np.tan(np.arange(0, np.pi, np.pi/36))
fig, axs = plt.subplots(1, 1, figsize=(LARGURA, ALTURA))
lim = [-10, 10]
for m in tangentes:
    axs.plot([lim[0]*m, lim[1]*m], [lim[0], lim[1]], linestyle='-')
axs.plot(0, 0, color=P_COLOR, marker='o')
axs.set_xlabel('$X$')
axs.set_ylabel('$Y$')
axs.set_xlim(-5, 5)
axs.set_ylim(-5, 5)
axs.set_xticklabels([])
axs.set_yticklabels([])
plt.savefig(f'img/{nome_img}.png', dpi=N_DPI) if SALVAR_GRAFICOS else None
```

```python
nome_img = 'dois_pontos'

fig, axs = plt.subplots(1, 1, figsize=(LARGURA, ALTURA))
lim = [-10, 10]
axs.plot([lim[0], lim[1]], [lim[0], lim[1]], linestyle='-')
axs.plot(-2, -2, color=P_COLOR, marker='o')
axs.plot(2, 2, color=P_COLOR, marker='o')
axs.set_xlabel('$X$')
axs.set_ylabel('$Y$')
axs.set_xlim(-5, 5)
axs.set_ylim(-5, 5)
axs.set_xticklabels([])
axs.set_yticklabels([])
plt.savefig(f'img/{nome_img}.png', dpi=N_DPI) if SALVAR_GRAFICOS else None
```

```python
nome_pasta = 'plano_xy'
cria_pasta(nome_pasta)

ms = [1, 0.9, 0.8, 0.7, 0.6]
ns = [0, 1, 2, 3, 4]
x1 = 0
x2 = 20
for i in range(0, len(ms)+1):
    fig, axs = plt.subplots(1, 1, figsize=(LARGURA, ALTURA))
    plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
    for j in range(0,i):
        if j == 0:
            axs.plot([x1,x2], [x1*ms[j]+ns[j],x2*ms[j]+ns[j]], color=COLORS[j], label=f'$y=x$')
        else:
            axs.plot([x1,x2], [x1*ms[j]+ns[j],x2*ms[j]+ns[j]], color=COLORS[j], label=f'$y={ms[j]}x+{ns[j]}$')
    axs.plot(10, 10,'o', color=P_COLOR, label='')
    axs.set_xlabel('$X$')
    axs.set_ylabel('$Y$')
    axs.legend()
    axs.set_xlim(x1,x2)
    axs.set_ylim(x1,x2)
    plt.savefig(f'img/{nome_pasta}/{nome_pasta}{i}.png', dpi=N_DPI) if SALVAR_GRAFICOS else None
```
