```python
import random
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')
plt.rcParams['axes.facecolor'] = '#dfdfea'
```

```python
salvar_graficos = 1

nome_pasta = 'votacao'
n_dpi = 240
p = (10, 10)
p_color = (0, 0, 0)

# Coordenadas dos pontos originais
xs = [0.6, 0.7, 0.8, 0.9, 1]
ys = [0, 1, 2, 3, 4]
colors = [(24/255, 192/255, 196/255), (246/255, 33/255, 150/255), (162/255, 103/255, 245/255), (243/255, 144/255, 126/255), (255/255, 228/255, 107/255)]

n_img = 5
largura = 14/2.54
altura = 10/2.54

fig, axs = plt.subplots(1, 1, figsize=(largura, altura))
for i in range(0, n_img):
    axs.plot(xs[i], ys[i], 'o', color=colors[i], label=f'$({xs[i]};{ys[i]})$')
axs.set_xlabel('x')
axs.set_ylabel('y')

# Inverter o eixo x
axs.invert_xaxis()

# Ajustar os limites do eixo x e y para cobrir todos os pontos
axs.set_xlim(min(xs) - 0.02, max(xs) + 0.02)
axs.set_ylim(min(ys) - 0.2, max(ys) + 0.2)  # Aumentar o eixo y

plt.savefig(f'img/{nome_pasta}/{nome_pasta}{111}.png', dpi=n_dpi) if salvar_graficos else None

```

```python
nome_pasta = 'votacao'
p = (10, 10)
p_color = (0, 0, 0)

a = 9.0
b = -5.0

# Número de pontos desejados
num_points = 35

# Pontos iniciais
xs = [0.6, 0.7, 0.8, 0.9, 1, 0.65, 0.65, 0.7, 0.75, 0.75, 0.85, 0.85, 0.9, 1, 0.95]
ys = [0, 1, 2, 3, 4, 0.1, 1.2, 2.5, 0.5, 2.5, 3, 2, 4, 3.5, 2]

# Gere pontos adicionais com dispersão aleatória
for _ in range(num_points):
    x = random.uniform(min(xs), max(xs))
    y = a * x + b + random.uniform(-1.5, 1.5)  # Adiciona ruído aos valores y
    xs.append(x)
    ys.append(y)

colors = [(0, 0, 0), (0, 0, 0), (162/255, 103/255, 245/255), (0, 0, 0), (0, 0, 0)]

# Gerar 10 pontos aleatórios e adicioná-los às listas existentes
for _ in range(50):
    new_color = (0, 0, 0)  # Gerar uma cor aleatória
    colors.append(new_color)

n_img = len(xs)  # Atualizar o número total de pontos

largura = 14/2.54
altura = 10/2.54

fig, axs = plt.subplots(1, 1, figsize=(largura, altura))

# Plotar os pontos existentes como marcadores individuais
for i in range(n_img):
    axs.plot(xs[i], ys[i], 'o', color=colors[i])



axs.set_xlabel('x')
axs.set_ylabel('y')

# Inverter o eixo x
axs.invert_xaxis()

# Ajustar os limites do eixo x e y para cobrir todos os pontos
axs.set_xlim(min(xs) - 0.02, max(xs) + 0.02)
axs.set_ylim(min(ys) - 0.2, max(ys) + 0.2)  # Aumentar o eixo y

plt.savefig(f'img/{nome_pasta}/{nome_pasta}{222}.png', dpi=n_dpi) if salvar_graficos else None
```
