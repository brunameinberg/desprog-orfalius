```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')
```

```python
n_dpi = 480
p = (10,10)
p_color = (0,0,0)
n_img = 1

x1 = 0
x2 = 20

ms = [1, 0.9, 0.8, 0.7, 0.6]
ns = [0, 1, 2, 3, 4]
colors = [(24/255, 192/255, 196/255), (246/255, 33/255, 150/255), (162/255, 103/255, 245/255), (243/255, 144/255, 126/255), (255/255, 228/255, 107/255)]

largura = 28/2.54
altura = 10/2.54
```

```python
fig, axs = plt.subplots(1, 2, figsize=(largura, altura))

# plt.figure(figsize=(14/2.54*2, 10/2.54*2))
for i in range(0,5):
    axs[0].plot([x1,x2], [x1*ms[i]+ns[i],x2*ms[i]+ns[i]], color=colors[i], label=f'$y={ms[i]}x+{ns[i]}$')
axs[0].plot(p[0], p[1],'o', color=p_color)
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')
axs[0].legend()
axs[0].set_xlim(x1, x2)
axs[0].set_ylim(x1, x2)


# plt.figure(figsize=(14/2.54*2, 10/2.54*2))
for i in range(0,n_img):
    axs[1].plot(ms[i],ns[i] ,'o' , color=colors[i], label=f'$m={ms[i]};n={ns[i]}$')
# axs[1].plot(p[0], p[1],'o', color=(24/255, 192/255, 196/255))
axs[1].set_xlabel('m')
axs[1].set_ylabel('n')
axs[1].legend()
axs[1].set_xlim(ms[-1]-0.02, ms[0]+0.02)
axs[1].set_ylim(ns[0]-0.2, ns[-1]+0.2)

plt.savefig(f'img/ani2/ani0{n_img}.png', dpi=n_dpi)
n_img += 1
```
