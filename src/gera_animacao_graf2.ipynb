```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')
```

```python
salvar_graficos = 1

nome_pasta = 'plano_mn'
n_dpi = 300
p = (10,10)
p_color = (0,0,0)
n_img = 1

x1 = 0
x2 = 20

ms = [1, 0.9, 0.8, 0.7, 0.6]
ns = [0, 1, 2, 3, 4]
colors = [(24/255, 192/255, 196/255), (246/255, 33/255, 150/255), (162/255, 103/255, 245/255), (243/255, 144/255, 126/255), (255/255, 228/255, 107/255)]

largura = 14/2.54
altura = 10/2.54
```

```python
fig, axs = plt.subplots(1, 1, figsize=(largura, altura))

for i in range(0,n_img):
    axs.plot(ms[i],ns[i] ,'o' , color=colors[i], label=f'$({ms[i]};{ns[i]})$')
axs.set_xlabel('m')
axs.set_ylabel('n')
axs.legend()
axs.set_xlim(ms[-1]-0.02, ms[0]+0.02)
axs.set_ylim(ns[0]-0.2, ns[-1]+0.2)

plt.savefig(f'img/{nome_pasta}/{nome_pasta}{n_img}.png', dpi=n_dpi) if salvar_graficos else None
n_img += 1
```
