```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')
```

```python
salvar_graficos = 0

nome_img = 'ex'
n_dpi = 300
p_color = (0,0,0)
n_img = 1

x1 = 0
x2 = 20

ms = [1, 0.9, 0.8, 0.7, 0.6]
ns = [0, 1, 2, 3, 4]
colors = [(24/255, 192/255, 196/255), (246/255, 33/255, 150/255), (162/255, 103/255, 245/255), (243/255, 144/255, 126/255), (255/255, 228/255, 107/255)]

largura = 14/2.54
altura = 14/2.54
```

```python
plt.figure(figsize=(largura, altura))
for i in range(0,1):
    plt.plot([x1,x2], [x1*ms[i]+ns[i],x2*ms[i]+ns[i]], color=colors[i], label=f'$y={ms[i]}x+{ns[i]}$')
plt.plot(10, 10,'o', color='orange')
plt.plot(15, 15,'o', color='blue')
plt.plot(5, 5,'o', color='darkgreen')
plt.xlabel('X')
plt.ylabel('Y')
# plt.legend()
plt.xlim(x1,x2)
plt.ylim(x1,x2)
plt.savefig(f'img/{nome_img}{n_img}.png', dpi=n_dpi) if salvar_graficos else None
n_img += 1
```

```python
fig, axs = plt.subplots(1, 1, figsize=(largura, altura))

axs.plot((ms[-1]-0.02, ms[0]+0.02), (ns[-1]+0.2, ns[0]-0.2), color=p_color)
# for i in range(0,n_img):
#     axs.plot(ms[i],ns[i] ,'o' , color=colors[i], label=f'$m={ms[i]};n={ns[i]}$')
axs.set_xlabel('m')
axs.set_ylabel('n')
# axs.legend()
# axs.set_xticks(np.arange(ms[-1]-0.02, ms[0]+0.02, 0.1))  # Intervalo de 2 unidades no eixo x
# axs.set_yticks(np.arange(ns[0]-0.2, ns[-1]+0.2, 1))  # Intervalo de 0.5 unidades no eixo y
axs.set_xticks(np.arange(ms[-1], ms[0]+0.1, 0.1))
axs.set_yticks(np.arange(ns[0], ns[-1]+0.2, 1))
axs.set_xlim(ms[-1]-0.02, ms[0]+0.02)
axs.set_ylim(ns[0]-0.2, ns[-1]+0.2)

plt.savefig(f'img/{nome_img}{n_img}.png', dpi=n_dpi) if salvar_graficos else None
n_img += 1
```
